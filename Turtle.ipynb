```python
import turtle
import random

# Set up the screen
screen = turtle.Screen()
screen.setup(500, 500)
screen.title("Turtle Game")

# Create the player turtle
player = turtle.Turtle()
player.shape("turtle")
player.color("green")
player.penup()

# Create the goal turtle
goal = turtle.Turtle()
goal.shape("circle")
goal.color("red")
goal.penup()
goal.goto(random.randint(-200, 200), random.randint(-200, 200))

# Define the player's movement functions
def move_up():
    player.setheading(90)
    player.forward(10)

def move_down():
    player.setheading(270)
    player.forward(10)

def move_left():
    player.setheading(180)
    player.forward(10)

def move_right():
    player.setheading(0)
    player.forward(10)

# Set up the keyboard bindings
screen.listen()
screen.onkey(move_up, "Up")
screen.onkey(move_down, "Down")
screen.onkey(move_left, "Left")
screen.onkey(move_right, "Right")

# Define the collision detection function
def is_collision(t1, t2):
    distance = t1.distance(t2)
    if distance < 20:
        return True
    else:
        return False

# Set up the game loop
while True:
    # Move the goal turtle randomly
    goal.setheading(random.randint(0, 360))
    goal.forward(5)

    # Check for collision between the player and goal turtles
    if is_collision(player, goal):
        goal.goto(random.randint(-200, 200), random.randint(-200, 200))

    # Update the screen
    screen.update()
```
